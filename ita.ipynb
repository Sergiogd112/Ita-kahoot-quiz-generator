```python
import pandas as pd
import numpy as np
import os
import PyPDF2
from pprint import pprint
import re
import unicodedata
```

```python
finalex_path="Final exam"
files=list(os.listdir(finalex_path))[::-1]
```

```python
file =files[0]
fpath=os.path.join(finalex_path,file)
print(fpath)
text=""
with open(fpath,'rb') as f:
    pdfreader=PyPDF2.PdfFileReader(f)
    x=pdfreader.numPages
    for pn in range(x):
        pageobj=pdfreader.getPage(pn)
        text+=pageobj.extractText()+'\n'
```

```python
testsplit = "INFRAESTRUCTURES DEL TRANSPORT AERI (ITA)"
test = (
    (
        "1."
        + "1.".join(
            text.split("INFRAESTRUCTURES DEL TRANSPORT AERI (ITA)")[1].split("1.")[1:]
        )
    )
    .replace("\nigth", "fligth")
    .replace("ffligth", "fligth")
)
with open("test.txt", "w",encoding="utf8") as f:
    f.write(test)
arrdict = []
qcount = 0
regex = r"^(\d{1,2}\.)+\d{1,2}"
qpattern = re.compile("^(\d{1,2}\.)+\d{1,2}")
ccount = 0
main=""
for n, line in enumerate(test.split("\n")[:-1]):
    line = line.strip()
    if ccount == 0 and line.split(". ")[0].isdigit() and len(line.split(". ")) > 1:
        op = "q"
        arrdict += [{"q": line}]
        main+=line+'\n'
        ccount = 1
    elif line[:3] in "(a)(b)(c)(d)":
        main+=line[3:]+'\n'
        op = line[:3]
        arrdict[-1][op] = line[3:]
        ccount = (ccount + 1) % 5
    else:
        
            print(ccount, line[:3] in "(a)(b)(c)(d)", line[:3])
            print(n, line, op)
            arrdict[-1][op] += " " + line
            main=main[:-1]+" "+line+'\n'
```

```python
pprint(main)
with open("main.txt",'w') as f:
    f.write(main)
```
